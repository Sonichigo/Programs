```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


plt.rcParams['figure.figsize'] = (12.0, 9.0)
df1 = pd.read_csv('C:\\Users\\KIIT\\Downloads\\Finance_data.csv')
df1.describe()

scaler = MinMaxScaler()
df_norm = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
print("Normal Data")
print(df_norm)
```

```python
X = df1['rm']
Y = df1['medv']

m=0
c=0
L=0.0001
epochs=1000
n=float(len(X))
list_m=[]
list_c=[]

for i in range(epochs):
    Y_pred=(m*X) +c
    cost=(1/n)*sum([i**2 for i in (Y-Y_pred)])
    D_m= (-2/n)*sum(X*(Y-Y_pred))
    D_c= (-2/n)*sum(Y-Y_pred)
    m=m-L*D_m
    c=c-L*D_c
    list_m.append(D_m)
    list_c.append(D_c)

print(m)
print(c)
print(cost)
```

```python
Y_pred = m*X + c

plt.scatter(X,Y)
plt.xlabel('iteration')
plt.ylabel('m')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red',linewidth=3) 
plt.show()
```
