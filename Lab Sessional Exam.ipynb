```python
#1. Implement Stack in Python.Along with Push and Pop Operations.

def crt_stack():
    stacp = []
    return stacp

def check_empty(stack):
    return len(stack) == 0

def push(stack, item):
    stack.append(item)
    print("pushed: " + item)

def pop(stack):
    if (check_empty(stack)):
        return "empty"
    return stack.pop()

num = input()
num1 = input()
num2 = input()
num3 = input()

stack = crt_stack()
push(stack, str(num))
push(stack, str(num1))
push(stack, str(num2))
push(stack, str(num3))
print("before stack: " + str(stack))
print("popped: " + pop(stack))
print("after stack: " + str(stack))
```

```python
#2 Download the Titanic Dataset and perform classification using Logistic regression and decision tree.
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

plt.rcParams['figure.figsize'] = (12.0, 9.0)
df1 = pd.read_csv('C:\\Users\\KIIT\\Downloads\\titanic.csv', sep='\t')
df1.describe()
```

```python
X = df1.iloc[:, 0:1].values
Y = df1.iloc[:, 1].values 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)


# Logistic Regression
classifier3 = LogisticRegression(solver='lbfgs', max_iter=1000)
classifier3.fit(X_train,Y_train)

p_model = classifier3.predict(X_test)
score = classifier3.score(X_test, Y_test.astype(int))
cm = metrics.confusion_matrix(Y_test.astype(int), p_model)

plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);

y_train_predict2 = classifier3.predict(X_train)
rmse2=(np.sqrt(mean_squared_error(Y_train.astype(int),y_train_predict2)))
r2_2=r2_score(Y_train.astype(int),y_train_predict2)
print('RMSE :',rmse2,'R2_SCORE: ',r2_2)
```

```python
#Decision Tree

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
Y_pred=model.predict(X_test)
accuracy=metrics.accuracy_score(Y_test,Y_pred)
print(accuracy)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```
